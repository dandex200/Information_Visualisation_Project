```python
#Import Pandas
import pandas as pd
```

The chosen dataset for this project was the first option in the 2018 Tidy Tuesday's Datasets. It can be found here: https://github.com/rfordatascience/tidytuesday/blob/master/data/2018/2018-04-02/us_avg_tuition.xlsx?raw=true

In order to apply pre-processing in pandas to this dataset, I needed to read it into a dataframe as an excel document then convert it to a CSV, then read that CSV file into a dataframe. 

```python
df = pd.read_excel("us_avg_tuition.xlsx")
```

```python
df.to_csv("us_avg_tuition.csv")
```

```python
df = pd.read_csv("us_avg_tuition.csv", index_col=False)
```

```python
df.head(10)
```

As we can see above, there is an additional id column that we do not need so we will remove that. 

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df.set_index("State")
```

We now need to add a column that contains the average of each row. 

```python
df['Average'] = df.mean(axis=1)
```

To shorten the large values for each column, we will turn each value into an int. 

```python
df = df.astype({"2004-05": "int", "2005-06": "int", "2006-07": "int", "  2007-08 ": "int", "2008-09": "int", "2009-10": "int", "2010-11": "int", "2011-12": "int", "2012-13": "int", "2013-14": "int", "2014-15": "int", "2015-16": "int", "Average": "int"})
```

```python
df.head(10)
```

```python
df.to_csv("us_avg_tuition.csv")
```
